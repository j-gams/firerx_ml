```python
### reduce with "valid" padding
def cv(dim, k, s):
    res = math.floor((dim - k) / s) + 1

### reduce with "same" padding
def ca(dim, s):
    res = math.floor((dim - 1) / s) + 1
    return res

### expand with "valid" padding
def dv(dim, k, s):
    res = ((dim - 1) * s) + k
    return res

### expand with "same" padding
def da(dim, s):
    res = dim * s
    return res
```

```python
class divtree:
    def __init__(self, dimval):
        self.left = None
        self.center = None
        self.right = None
        self.depth = None
        self.parent = None
        self.input = None
        self.outdim = dimval
        self.prevops = []
```

```python
def treecurse(centernode, inputs, depth):
    ### divide into lrc
    r = []
    c = []
    l = []
    rrank = (float("inf"), None)
    lrank = (float("-inf"), None)
    for elt in inputs:
        if elt.outdim < centernode.outdim:
            if elt.outdim > lrank[0]:
                lrank = (elt.outdim, len(l))
            l.append(elt)
        elif elt.outdim == centernode.outdim:
            elt.parent = centernode
            elt.depth = depth + 1
            centernode.center = elt
            c.append(elt)
        else:
            if elt.outdim < rrank[0]:
                rrank = (elt.outdim, len(r))
            r.append(elt)
    print(centernode.outdim, rrank, lrank)
    if len(l) > 0:
        centernode.left = l[lrank[1]]
        treecurse(l[lrank[1]], l, depth+1)
    if len(r) > 0:
        centernode.right = r[rrank[1]]
        treecurse(r[rrank[1]], r, depth+1)
```
